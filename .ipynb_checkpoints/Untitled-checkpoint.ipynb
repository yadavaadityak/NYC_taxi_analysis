```python
import pandas as pd  
import pyspark.pandas as ps 
from pyspark.sql import SparkSession 

# this is define schema before importing files
from pyspark.sql.types import * 
from pyspark.sql.functions import col  

import pyarrow.parquet as pq
import pyarrow as pa


import pathlib


# pg 52 from Learning Spark 
spark = SparkSession.builder.getOrCreate()
```

```python
def data_processing(df):
    """
    Performs action on data to  
    """

    # for NYC yellow taxi files
    if "tpep_pickup_datetime" in df.columns:
        df = df.withColumn("pickup_timestamp", col("tpep_pickup_datetime").cast("timestamp"))
        df = df.withColumn("dropoff_timestamp", col("tpep_dropoff_datetime").cast("timestamp"))

        # Calculate trip duration (in minutes)
        df = df.withColumn(
            "trip_duration_min",
            (col("dropoff_timestamp").cast("long") - col("pickup_timestamp").cast("long")) / 60.0
        )
    
    # for Uber+ files 
    if "trip_time" in df.columns: 
        df = df.withColumn("trip_time", col("trip_time") / 60)
        df = df.withColumn("total_fare", col("base_passenger_fare") + col("tolls") + col("bcf") + 
                           col("sales_tax") + col("congestion_surcharge") + col("airport_fee"))
        df = df.withColumn("company_commission", df["total_fare"] - df["driver_pay"])

    return df 
```
